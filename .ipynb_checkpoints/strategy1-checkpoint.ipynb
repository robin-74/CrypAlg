```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
class CryptoData:
    def add_class(self, ticker, start, end):
        self.ticker = ticker
        self.start = start
        self.end = end
        self.data = self.init_data()
    
    def plot_data(self, df):
        plt.figure(figsize=(12, 6))
        sns.lineplot(x=df.index, y=df["Price"].squeeze(), label="Price")
        plt.title(f"{self.ticker} Price")
        plt.xlabel("Date")
        plt.ylabel("Price")
        plt.legend()
        plt.show()

    def init_data(self):
        data1 = yf.download(self.ticker, self.start, self.end)
        df = data1[['Close']].copy()
        df.rename(columns={"Close": "Price"}, inplace=True)
        self.plot_data(df)
        return df

    def get_data(self):
        return self.data

    def set_sma(self):
        list_sma = []
        n = int(input("Enter how many SMA you want: "))
        for i in range(n):
            window = int(input(f"Enter SMA window size for SMA #{i+1}: "))
            list_sma.append(window)
        for window in list_sma:
            col_name = f"SMA_{window}"
            self.data[col_name] = self.data["Price"].rolling(window=window).mean()
        plt.figure(figsize=(12, 6))
        sns.lineplot(x=self.data.index, y=self.data["Price"].squeeze(), label="Price")
        for window in list_sma:
            col_name = f"SMA_{window}"
            sns.lineplot(x=self.data.index, y=self.data[col_name].squeeze(), label=col_name)
        plt.title(f"{self.ticker} Price with SMA(s)")
        plt.xlabel("Date")
        plt.ylabel("Price")
        plt.legend()
        plt.show()
```

```python
data_class = CryptoData()
data_class.add_class("BTC-USD", "2024-01-01", "2025-04-04")
```
