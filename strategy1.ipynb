```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

class CryptoData:
    def add_class(self, ticker, start, end):
        self.ticker = ticker
        self.start = start
        self.end = end
        self.data = self.get_data()

    def plot_data(self, df):
        # Plot the price data
        plt.figure(figsize=(12, 6))
        sns.lineplot(x=df.index, y=df["Price"])
        plt.show()

    def get_data(self):
        data1 = yf.download(self.ticker, self.start, self.end)
        data2 = data1["Close"]
        if isinstance(data2, pd.DataFrame):
            data3 = data2["BTC-USD"]
        df = data3.to_frame(name="Close")
        df.rename(columns={"Close": "Price"}, inplace=True)
        self.plot_data(df)
        return df
crypto = CryptoData()
crypto.add_class("BTC-USD", "2024-01-01", "2025-04-04")
```
