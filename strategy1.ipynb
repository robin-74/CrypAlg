```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

class CryptoData:
    def add_class(self, ticker, start, end):
        self.ticker = ticker
        self.start = start
        self.end = end
        self.data = self.init_data()

    def plot_data(self, df):
        plt.figure(figsize=(12, 6))
        sns.lineplot(x=df.index, y=df["Price"], label="Price")
        plt.title(f"{self.ticker} Price")
        plt.xlabel("Date")
        plt.ylabel("Price")
        plt.legend()
        plt.show()

    def init_data(self):
        data1 = yf.download(self.ticker, self.start, self.end)
        # Select the 'Close' column and rename it to 'Price'
        df = data1[['Close']].copy()
        df.rename(columns={"Close": "Price"}, inplace=True)
        self.plot_data(df)
        return df

    def get_data(self):
        return self.data

    def set_sma(self):
        list_sma = []
        n = int(input("Enter how many SMA you want: "))
        for i in range(n):
            window = int(input(f"Enter SMA window size for SMA #{i+1}: "))
            list_sma.append(window)
        # Calculate the SMAs and append them to self.data
        for window in list_sma:
            col_name = f"SMA_{window}"
            self.data[col_name] = self.data["Price"].rolling(window=window).mean()
        # Plot the Price data along with the SMA lines
        plt.figure(figsize=(12, 6))
        sns.lineplot(x=self.data.index, y=self.data["Price"], label="Price")
        for window in list_sma:
            col_name = f"SMA_{window}"
            sns.lineplot(x=self.data.index, y=self.data[col_name], label=col_name)
        plt.title(f"{self.ticker} Price with SMA(s)")
        plt.xlabel("Date")
        plt.ylabel("Price")
        plt.legend()
        plt.show()

# Example usage:
crypto = CryptoData()
crypto.add_class("BTC-USD", "2024-01-01", "2025-04-04")
# This will prompt for SMA window sizes, calculate them, and replot the chart with SMA lines.
crypto.set_sma()
```
